### Q2.

```python
import numpy as np
import pandas as pd
series_A = pd . Series([ 1 , 2 , 3 , 4 , 5 ])
series_B = pd . Series([ 4 , 5 , 6 , 7 , 8 ])

#union of the two series
union = pd.Series(np.union1d(series_A, series_B)) 
# print(union)

# Q2.b
#intersection of the two series
common = pd.Series(np.intersect1d(series_A, series_B))
print("Items common to both series A and series B")
print(list(common))

# Q2.a
#not common elements of the two series.
notcommon = union[~union.isin(common)]
print("Items not common to both series A and series B")
print(list(notcommon))
```

### Q5.

```python
#creating Train dataframe.
train = pd.DataFrame({'State':['A','B','C','A','A','B','A','A','C','C'],
                   'class':[0,1,1,0,1,1,0,1,1,0]
                  })
# Train Data
print(train)


# Response Table
response=train.groupby('State')['class'].value_counts(normalize=True).to_frame().unstack().reset_index()
response.columns = response.columns.droplevel(1)
response.columns = ['State','Class_0','Class_1']
response.fillna(0,inplace=True)
print('\n',response)


# Encoded Train Data
encode = pd.merge(train,response,how='left')
encode.drop(columns=['State','class'],inplace=True)
encode.rename(columns={'Class_0':'State_0','Class_1':'State_1'})
encode
```
